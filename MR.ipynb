```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movies2.csv")
```

```python
print(df.columns.values)
```

```python
features = ['genre_names', 'popularity', 'original_title']
```

```python
df['popularity'].isnull().values.any()
```

```python
def combine_features(row):
    return (
        str(row['original_title']) + ' ' +
        str(row['genre_names']) + ' ' +
        str(row['overview']) + ' ' +
        str(row['popularity']) + ' ' +
        str(row['id'])
    )
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
columns_to_combine = ['original_title', 'genre_names', 'overview', 'popularity', 'id']
```

```python
for col in columns_to_combine:
    if col in df.columns:
        df[col] = df[col].fillna('') 
```

```python
print(df.dtypes)
```

```python
df['combine_features'] = df.apply(combine_features, axis=1)
```

```python
print(df.loc[5, 'combine_features'])
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['combine_features'])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["original_title"].values[0]
def get_index_from_original_title(original_title):
    return df[df.original_title == original_title]["index"].values[0]
```

```python
def get_index_from_original_title(movie_title):
    try:
        return df[df['original_title'] == movie_title].index[0]
    except IndexError:
        print(f"Error: Movie title '{movie_title}' not found.")
        return None
```

```python
print(df['original_title'].unique()) 
```

```python

def get_index_from_original_title(movie_title):
    try:
        return df[df['original_title'] == movie_title].index[0]
    except IndexError:
        print(f"Error: Movie title '{movie_title}' not found.")
        return None

movie_user_likes = "Pulp Fiction"

movie_index = get_index_from_original_title(movie_user_likes)

if movie_index is not None:
    similar_movies = list(enumerate(cosine_sim[movie_index]))
    print(similar_movies)
else:
    print("Cannot find similar movies as the movie was not found.")
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python

def get_original_title_from_index(index):
    try:
        return df.loc[index, 'original_title']
    except KeyError:
        print(f"Error: Index {index} is not valid.")
        return None

print("Top 20 similar movies to " + movie_user_likes + " are:\n")
i = 0
for element in sorted_similar_movies:
    print(get_original_title_from_index(element[0]))  
    i += 1
    if i > 20:
        break
```
